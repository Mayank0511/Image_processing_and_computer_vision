Image processing basics

Adding and subtracting two images using different methods and comparing their processing times.

```python
#method 1: list of comprehension method(or loc method)
import numpy as np
import matplotlib.pyplot as plt
import time
import cv2
def imgadder(img1,img2):                                      
    x,y,z=img1.shape
    img3=[img1[i]+img2[i] for i in range(x)]   #list comprehension expression
    img3= np.asarray(img3, dtype=np.uint8)
    new = np.reshape(img3, [x,y,z])    
    return new
def imgsubtractor(img1,img2):                                      
    x,y,z=img1.shape
    img3=[img2[i]-img1[i] for i in range(x)]    #list comprehension expression
    img3= np.asarray(img3, dtype=np.uint8)
    new = np.reshape(img3, [x,y,z])    
    return new

img1= cv2.imread('C:\\Users\mayan\Downloads\input1.jpg')
img2= cv2.imread('C:\\Users\mayan\Downloads\input2.jpg')   #Importing the images
plt.imshow(img1)
plt.title('Input image 1:')
plt.show()
plt.imshow(img2)
plt.title('Input image 2:')
plt.show()

t21=time.time()
added_img=imgadder(img1,img2)
t22=time.time()
t_locadd=t22-t21
print("time taken by loc method to add two images: ",t_locadd)
plt.imshow(added_img)
plt.title('output added image:loc method')
plt.show()
t31=time.time()
sub_img=imgsubtractor(img1,img2)
t32=time.time()
t_locsub=t22-t21
print("time taken by loc method to subtract two images: ",t_locsub)
plt.imshow(sub_img)
plt.title('output subtracted image:loc method')
plt.show()


#method 2: cv2 method

image3=cv2.imread('C:\\Users\mayan\Downloads\input1.jpg')
image4=cv2.imread('C:\\Users\mayan\Downloads\input2.jpg')
t41=time.time()
image5=cv2.add(image3,image4)
t42=time.time()
t_addcv2=t42-t41
print("time taken by cv2method to add two images: ",t_addcv2)
plt.imshow(image5)
plt.title('output subtracted image:cv2 method')
plt.show()

#using Inbuilt function
t51=time.time()
image6=cv2.addWeighted(image4,-1,image3,1,0)
t52=time.time()
t_cv2sub=t52-t51
print("time taken by loc method to subtract two images: ",t_cv2sub)
plt.imshow(image6)
plt.title('output subtracted image: using inbuilt function')
plt.show()

#Comparing processing times of different methods for add and sub
x=['loc add','loc sub','cv2 add','cv2 sub']
process_times=[t_locadd,t_locsub,t_addcv2,t_cv2sub]
x_pos=[i for i in range(len(x))]
plt.bar(x_pos,process_times,color='blue')
plt.xlabel('   <------METHOD------>   ')
plt.ylabel('   <-----processing times----->   ')
plt.title('processing times vs methods')
plt.xticks(x_pos,x)
plt.show()
```
